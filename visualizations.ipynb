# t-SNE visualizations

```python
import pandas as pd
import pickle
import break_utils
```

```python
def create_tsne(weights_name, layer):
    df = break_utils.load_reps(weights_name)
    meaning_types = list(df.meaning.value_counts().index)
    df['meaning_color'] = df['meaning'].apply(lambda x: break_utils.get_color(x, meaning_types))
    break_utils.tsne_viz(df, weights_name, f'break_rep_layer{layer}')
```

```python
def create_all_tsne(weights_name):
    if 'large' in weights_name:
        layers = (1, 6, 12, 18, 24)
    else:
        layers = (1, 6, 12)
    for layer in layers:
        create_tsne(weights_name, layer)
```
